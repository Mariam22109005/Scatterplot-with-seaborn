<span style="color:black;"> Firstly import seaborn library .Load the default iris dataset then use regplot function which is mostly used as scatter plot visualization.The **regplot() function** of the Seaborn library can be used to create a scatterplot. In the example, the iris dataset, a sample dataset from the Seaborn repository, is used. The graph illustrates the association between plant breadth and sepal length. The following inputs must be given in order to demonstrate how this function is used in its simplest form:

* x: the places of points on the X axis
    
* y: the places of points on the Y axis
The graphic is by default a linear regression fit. By setting fit reg=False, it can be eliminated from the graph.</span>




<div style="background-color:#d4878e; color:black; padding:10px; border-radius:10px; text-align:center;">
    <h3 style="font-weight:bold; margin-top:0; font: 30px Arial;"># Representation  of simple scatterplot example </h3>
</div>

```python
# library & dataset
import seaborn as sns
df = sns.load_dataset('iris')

# use the function regplot to make a scatterplot
sns.regplot(x=df["sepal_length"], y=df["sepal_width"])
```

<div style="background-color:#d4878e; color:black; padding:10px; border-radius:10px; text-align:center;">
    <h3 style="font-weight:bold; margin-top:0; font: 30px Arial;"># Representation using color,transparency and width</h3>
</div>


With a scatterplot created with Seaborn, anyone  can customize how the regression fit looks.

In this illustration, the  line_kws={} option with the following components controls the color, transparency, and width:

* hue: the hue of the line
* alpha: the line's opacity value
* lw: the line's width

```python
# library &amp; dataset
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
 
# plot
sns.regplot(x=df["sepal_length"], y=df["sepal_width"], line_kws={"color":"r","alpha":0.7,"lw":5})
plt.show()
```


<div style="background-color:#d4878e; color:black; padding:10px; border-radius:10px; text-align:center;">
    <h3 style="font-weight:bold; margin-top:0; font: 30px Arial;"># Representation using marker  and fit_reg is false .</h3>
</div>


```python
# library and dataset
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
 
# change shape of marker
sns.regplot(x=df["sepal_length"], y=df["sepal_width"], marker="v", fit_reg=False)
plt.show()
```

<div style="background-color:#d4878e; color:black; padding:10px; border-radius:10px; text-align:left;">
    <h3 style="font-weight:bold; margin-top:0; font: 30px Arial;"># Representation of color dots according to group</h3>
</div>



This example uses **lmplot() function** of seaborn library. In order to define each species with different colors, species column of the dataset given in hue argument. The list of arguments needed for the function is:

* x : positions of points on the X axis
* y : positions of points on the Y axis
* data : dataset
* fit_reg : if True, show the linear regression fit line
* hue : variables that define subsets of the data
* legend : if True, add a legend
Note that the legend is specified in through matplotlib, instead of seaborn itself. In order to specifically define a location of the legend, plt.legend() can be used.

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
 
# Use the 'hue' argument to provide a factor variable
sns.lmplot( x="sepal_length", y="sepal_width", data=df, fit_reg=False, hue='species', legend=False)
 
# Move the legend to an empty part of the plot
plt.legend(loc='lower right')

plt.show()
```

<div style="background-color:#d4878e; color:black; padding:10px; border-radius:10px; text-align:center;">
    <h3 style="font-weight:bold; margin-top:0; font: 30px Arial;"># Representaton of color dots using marker for different group </h3>
</div>

 

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
 
# give a list to the marker argument
sns.lmplot( x="sepal_length", y="sepal_width", data=df, fit_reg=False, hue='species', legend=False, markers=["v", "x", "1"])
 
# Move the legend to an empty part of the plot
plt.legend(loc='lower right')

plt.show()
```

<div style="background-color:#d4878e; color:black; padding:10px; border-radius:10px; text-align:left;">
    <h3 style="font-weight:bold; margin-top:0; font: 30px Arial;"># Representation of different color which is defined by user</h3>

anyone can provide their preferred color palette by using the palette argument rather than utilizing the default color palette. Seaborn offers a wide range of color schemes, including **deep, muted, brilliant, pastel, dark, and colorblind**.

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
 
# Use the 'palette' argument
sns.lmplot( x="sepal_length", y="sepal_width", data=df, fit_reg=False, hue='species', legend=False, palette="Set2")
 
# Move the legend to an empty part of the plot
plt.legend(loc='lower right')
 
plt.show()
```

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
 
# Provide a dictionary to the palette argument
sns.lmplot( x="sepal_length", y="sepal_width", data=df, fit_reg=False, hue='species', legend=False, palette=dict(setosa="#f25e8b", virginica="#a69ef7", versicolor="#42e372"))
 
# Move the legend to an empty part of the plot
plt.legend(loc='lower right')
 
plt.show()
```

<div style="background-color:#d4878e; color:black; padding:10px; border-radius:10px; text-align:left;">
    <h3 style="font-weight:bold; margin-top:0; font: 30px Arial;"># Representation of scatterplot using axis limit </h3>
</div>

You can control the limits of X and Y axis of your plots using matplotlib **function plt.xlim() and plt.ylim().** In this example, lmplot() function of seaborn is used to plot a basic scatterplot with iris dataset. There are 2 arguments in these functions;

* plt.xlim() : (left, right)

* plt.ylim() : (buttom, top)

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
 
# basic scatterplot
sns.lmplot( x="sepal_length", y="sepal_width", data=df, fit_reg=False)
 
# control x and y limits
plt.ylim(0, 20)
plt.xlim(0, None)
 
plt.show()
```

<div style="background-color:#d4878e; color:black; padding:10px; border-radius:10px; text-align:left;">
    <h3 style="font-weight:bold; margin-top:0; font: 30px Arial;"># Representation of marginal plot  </h3>
</div>

A marginal plot is another type of scatterplot .A marginal plot enables the investigation of the connection between two numerical variables. Their association is seen in the central graph. **A scatterplot, hexbin plot, 2D histogram, or 2D density** plot is typically used. The marginal charts, which are often on the top and right, use a histogram or density plot to display the distribution of two variables.

To create these kinds of visuals, the Seaborn library has a combined plot function. Here are a few potential adjustments, and the top graph displays its default behavior.

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
 
# Custom the inside plot: options are: “scatter” | “reg” | “resid” | “kde” | “hex”
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='scatter')
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='hex')
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='kde')
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='resid')
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='reg')
plt.show()
```

```python
# Then you can pass arguments to each type:
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='scatter', s=200, color='m', edgecolor="skyblue", linewidth=2)
 
# Custom the color
sns.set(style="white", color_codes=True)
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='kde', color="skyblue")

plt.show()
```

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
 
# Custom the histogram:
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='hex', marginal_kws=dict(bins=30, fill=True))
plt.show()
```

<div style="background-color:#d4878e; color:black; padding:10px; border-radius:10px; text-align:left;">
    <h3 style="font-weight:bold; margin-top:0; font: 30px Arial;"># Representation of Ratio of marginal plot </h3>
</div>

It is also possible to change the central plot-margin plats ratio and the space between the joint and marginal axes.

* space : space between the joint and marginal axes
* ratio : ratio of joint axes height to marginal axes height

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
 
# No space
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='kde', color="grey", space=0)
 
# Huge space
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='kde', color="grey", space=3)
 
# Make marginal bigger:
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='kde',ratio=1)

plt.show()
```

<div style="background-color:#d4878e; color:black; padding:10px; border-radius:10px; text-align:left;">
    <h5 style="font-weight:bold; margin-top:0; font: 30px Arial;"># Representation of Pairplot with correlation</h5>
</div>

The seaborn library allows to draw a correlation matrix through the pairplot() function. The parameters to create the example graphs are:

* data : dataframe
* kind : kind of plot to make (possible kinds are ‘scatter’, ‘kde’, ‘hist’, ‘reg’)

```python
# library & dataset
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('iris')
 
# with regression
sns.pairplot(df, kind="reg")
plt.show()
 
# without regression
sns.pairplot(df, kind="scatter")
plt.show()
```

In a graph drawn by pairplot() function of seaborn, you can control the marker features, colors and data groups by using additional parameters such as:

* hue : variables that define subsets of the data
* markers : a list of marker shapes
* palette : set of colors for mapping the hue variable
plot_kws : a dictionary of keyword arguments to modificate the plot

```python
# library & dataset
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('iris')
 
# left
sns.pairplot(df, kind="scatter", hue="species", markers=["o", "s", "D"], palette="Set2")
plt.show()
 
# right: you can give other arguments with plot_kws.
sns.pairplot(df, kind="scatter", hue="species", plot_kws=dict(s=80, edgecolor="white", linewidth=2.5))
plt.show()
```

<div style="background-color:#d4878e; color:black; padding:10px; border-radius:10px; text-align:left;">
    <h5 style="font-weight:bold; margin-top:0; font: 30px Arial;"># Representation of Pairplot with distribution</h5>
</div>

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
 
# Density
sns.pairplot(df, diag_kind="kde")
 
# Histogram
sns.pairplot(df, diag_kind="hist")
 
# You can custom it as a density plot or histogram so see the related sections
sns.pairplot(df, diag_kind="kde", diag_kws=dict(shade=True, bw_adjust=.05, vertical=False) )

plt.show()
```
